```python
def predict (b, X):
        intercept = np.ones((X.shape[0], 1))
        X = np.c_[X, intercept]
        z = np.dot(b, X.T) 
        sig = 1/(1 + np.exp(-z))
        predict = [0 if i <=0.5 else 1 for i in sig]
        
        return predict
    

def logistic_regression(X,Y):
    iterations=150
    iteration_cnt=0
    conver_tol=0.000000001
    b=np.zeros(X.shape[1]+1)
    b1=[]
    intercept_X=np.ones((X.shape[0],1))
    X=np.c_[X, intercept_X]
    while((iteration_cnt<=iterations)):
        iteration_cnt+=1
        b1=b
        z = np.dot(b, X.T) 
        prob = 1/(1 + np.exp(-z))
        gradient = np.dot((Y - prob),X)
        we = prob*(1-prob)
        xP = np.array([i*j for (i,j) in zip(X,we)])
        hessian = -1*np.dot(xP.T, X)
        b = b - (np.dot(np.linalg.inv(hessian),gradient))
        fin = np.linalg.norm(b-b1)
        if( fin <= conver_tol):
            break
    return b,iteration_cnt
```
